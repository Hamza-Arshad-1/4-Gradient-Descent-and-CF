```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#make function of gradient_descent
def gd(x,y):
    m_cur = b_cur = 0;
    iterations = 1000
    n = len(x)
    learning_rate = 0.08
    for i in range(iterations):
        yp = m_cur *x + b_cur
        cost =  (1/n) * sum([val**2 for val in (y-yp)])
        #now find partial derivative of m
        md = -(2/n)*sum(x*(y-yp))
        bd = -(2/n) * sum(y-yp)
        m_cur = m_cur - learning_rate * md
        b_cur = b_cur - learning_rate * bd
        
        #print values
        #print("m {}, b {}, cost {}, iteration {}" .format(m_cur, b_cur, cost, i))
        #plot values
        plt.plot(x, yp, color = 'green')
    pass
```

```python
#creating array of x, y and for making equation
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
#calling function
gd(x,y);
```
